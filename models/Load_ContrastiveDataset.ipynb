```python
import pickle
```

```python
class ContrastiveDataset:
    def __init__(self):
        self.Jet1 = self.Jet()
        self.Jet2 = self.Jet()
        self.label = []
        
    def __len__(self):
        return len(self.label)
        
    class Jet:
        def __init__(self):
            self.jet_feats = []
            self.trk_feats = []
            self.jet_label = []
```

```python
data = pickle.load( open( "ContrastiveDataset.pkl", "rb" ) )
```

```python
num = 100

# Example of how to access the data by looping over some samples
for sample in range(num):
    print(data.Jet1.jet_feats[sample]) # These are the features for jet1
    print(data.Jet2.jet_feats[sample]) # These are the features for jet2
    print(data.Jet1.trk_feats[sample]) # These are the trk features associated to jet1 (variable length!)
    print(data.Jet2.trk_feats[sample]) # These are the trk features associated to jet2 (variable length!)
    print(data.Jet1.jet_label[sample]) # This is the pileup fraction of jet1. Closer to 0 is HS. Closer to 1 is PU.
    print(data.Jet2.jet_label[sample]) # This is the pileup fraction of jet2. Closer to 0 is HS. Closer to 1 is PU.
    print(data.label[sample])          # This is the Contrastive label for the jet1-jet2 pair.
    print()
```
