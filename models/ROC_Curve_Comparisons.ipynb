```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_names = ["JVT_Results.np","BaselineNN_Results.np","AttnJVT_Results.np","AttnJVT_wtracks_Results.np"]

results = []

for file in file_names:
    with open(file, 'rb') as f:
        results.append(np.load(f))
```

```python
x1 = results[0][0]
y1 = results[0][1]

x2 = results[1][0]
y2 = results[1][1]

x3 = results[2][0]
y3 = results[2][1]

x4 = results[3][0]
y4 = results[3][1]
```

```python
plt.title("JVT ROC Curve")
plt.plot(x1,y1,color='b',label='JVT')
plt.plot(x2,y2,color='r',label='BaselineNN')
plt.plot(x3,y3,color='g',label='AttnJVT')
plt.plot(x4,y4,color='m',label='AttnJVT wTracks')

plt.ylabel('Fake Rate',loc='top')
plt.xlabel('Signal Efficiency',loc='right')
plt.yscale('log')
plt.legend(loc='upper left')
plt.xlim([0.8, 1])
plt.grid(which='both')
plt.show()
```
