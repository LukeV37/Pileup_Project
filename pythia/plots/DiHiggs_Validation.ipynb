```python
import uproot
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
```

```python
higgs = {}
with uproot.open("../output/dataset_diHiggs_mu60_NumEvents10k_MinJetpT25.root:fastjet") as f:
    higgs['pt'] = f['jet_pt'].array()
    higgs['m'] = f['jet_m'].array()
    higgs['label'] = f['jet_pufr_truth'].array()
filter_none = {}
with uproot.open("../output/dataset_4b_NOFILTER_mu60_NumEvents10k_MinJetpT25.root:fastjet") as f:
    filter_none['pt'] = f['jet_pt'].array()
filter_20 = {}
with uproot.open("../output/dataset_4b_FILTER20_mu60_NumEvents10k_MinJetpT25.root:fastjet") as f:
    filter_20['pt'] = f['jet_pt'].array()
filter_40 = {}
with uproot.open("../output/dataset_4b_FILTER40_mu60_NumEvents10k_MinJetpT25.root:fastjet") as f:
    filter_40['pt'] = f['jet_pt'].array()
filter_60 = {}
with uproot.open("../output/dataset_4b_FILTER60_mu60_NumEvents10k_MinJetpT25.root:fastjet") as f:
    filter_60['pt'] = f['jet_pt'].array()
```

```python
plt.title("Jet pT")
plt.hist(ak.ravel(higgs['pt']),label='DiHiggs',histtype='step',color='r',bins=30,range=(0,600))
plt.hist(ak.ravel(filter_none['pt']),label='4b (No Filter)',linestyle='--',histtype='step',color='m',bins=30,range=(0,600))
plt.hist(ak.ravel(filter_20['pt']),label='4b (pTb Min 20 GeV)',linestyle='--',histtype='step',color='b',bins=30,range=(0,600))
plt.hist(ak.ravel(filter_40['pt']),label='4b (pTb Min 40 GeV)',linestyle='--',histtype='step',color='c',bins=30,range=(0,600))
plt.hist(ak.ravel(filter_60['pt']),label='4b (pTb Min 60 GeV)',linestyle='--',histtype='step',color='k',bins=30,range=(0,600))
plt.yscale('log')
plt.legend()
plt.show()
```

```python
plt.title("Jet Mass")
plt.hist(ak.ravel(higgs['m'][higgs['label']<0.005]),label='DiHiggs',histtype='step',color='r',bins=30,range=(0,200))
#plt.hist(ak.ravel(higgs['m'][higgs['label']>0.5]),label='Pileup',histtype='step',color='b',bins=30,range=(0,200))
#plt.yscale('log')
plt.legend()
plt.show()
```
